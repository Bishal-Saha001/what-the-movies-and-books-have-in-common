We want to know if this sample is representative of the population. We compute the Kolmogorov-Smirnov statistic on the observed sample distribution and the alleged original distribution for the null hypothesis that the sample is drawn from the another distribution.


Notes from https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.ks_2samp.html:

Computes the Kolmogorov-Smirnov statistic on 2 samples.This is a two-sided test for the null hypothesis that 2 independent samples are drawn from the same continuous distribution.

1. This tests whether 2 samples are drawn from the same distribution. Note that, like in the case of the one-sample K-S test, the distribution is assumed to be continuous.

2. This is the two-sided test, one-sided tests are not implemented. The test uses the two-sided asymptotic Kolmogorov-Smirnov distribution.

3. If the K-S statistic is small or the p-value is high, then we cannot reject the hypothesis that the distributions of the two samples are the same.

```python
stats.ks_2samp(df_book_subset.average_rating, df_book_noDict_year.average_rating)
```

The **pearson correlation** between both variables in this sample is:

```python
def pearson_r(x, y):
    """Compute Pearson correlation coefficient between two arrays."""
    # Compute correlation matrix: corr_mat
    corr_mat = np.corrcoef(x,y)
    # Return entry [0,1]
    return corr_mat[0,1]
```

```python
p_obs = pearson_r(df_sample_noDict_year.average_rating, df_sample_noDict_year.publication_year)
print('Pearson coefficient for the sample distribution: {}'.format(p_obs))
```

Therefore, this 1% of data is enough representative from the population. Now, we apply pairs bootstrapping measuring the pearson correlation coefficient between the two interest variables (in this case, the average rating and publication year) to confirm that relationship found in the sample distribution is not a chance.

```python
np.random.seed(42)
def pairs_bootstrap(data1, data2, func, n):
    
    p = np.empty(n)
    
    for i in range(n):
        inds = np.arange(len(data1))
        bs_inds = np.random.choice(inds, 5000)
        bs_data1 = data1[bs_inds]
        bs_data2 = data2[bs_inds]
    
        p[i] = func(bs_data1, bs_data2)
    
    return p
```

```python
pearson = pairs_bootstrap(df_book_noDict_year.average_rating, \
                          df_book_noDict_year.publication_year, \
                          pearson_r, 1000)
```

```python
p_value = np.sum(pearson >= p_obs)/len(pearson)
p_value
```

```python
alpha = 0.05
if p_value >= alpha:
    print('Fail to reject the null hypothesis')
else:
    print('Reject the null hypothesis')
```

As we demostrate here, the pearson coefficient is not a chance. We can't reject the null hypothesis, using a confidence level of 0.05, because we got a p-value of 0.358.

Chi-square bootstrapping test

We apply sucesive chi-square to n trials of subsample randomly chosen:

```python
from scipy.stats import chisquare

def chisquare_func(data1, data2):
    
    data1.dropna(subset=['language_code'], inplace=True)
    data2.dropna(subset=['language_code'], inplace=True)
    
    data1 = data1.groupby('language_code').count().counter.sort_values(ascending=False).head(10)
    data2 = data2.groupby('language_code').count().counter.sort_values(ascending=False).head(10)
    chisq, p = chisquare(f_obs=data1.values, f_exp=data2.values)
    return p
```

```python
np.random.seed(42)
def chi_square_bootstrap(data_source, data1, func, n):
    
    p = np.empty(n)
    
    for i in range(n):
        # we extract subsamples with replace to valid the bootstrapping trial
        data2 = df_book_noDict_year.sample(frac=0.2, replace=True)
        p[i] = func(data1, data2)
    
    return p
```

```python
p = chi_square_bootstrap(df_book_noDict_year, df_sample_languages, chisquare_func, 100)
```

How many times the p-value is more than 0.05?

```python
p_value = np.sum(p >= 0.05)/len(p)
```

```python
p_value 
```

```python
alpha = 0.05
if p_value >= alpha:
    print('Fail to reject the null hypothesis')
else:
    print('Reject the null hypothesis')
```
